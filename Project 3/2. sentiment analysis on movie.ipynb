```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv("movie description.csv")
```

```python
def sentiment_detection(sentence):
  positive = False
  negative = False
  for i in ('Fateful', 'Epic', 'Touching', 'Intrepid', 'Thoughtful', 'Action-Packed','Lacklusture '):
    if sentence.count(i) > 0:
      positive = True
  for i in ('Lacklusture', 'terrible', 'Boring', 'Taut', 'kill'):
    if sentence.count(i) > 0:
      negative = True
  if positive == negative == False:
    return "neutral"
  elif positive == True != negative:
    return "positive"
  elif negative == True != positive:
    return 'negative'
  else:
    return "mixed"
```

```python
df1['Classified As'] = df1['description'].apply(lambda x: sentiment_detection(x))
df1
```
