```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

countries_df = pd.read_csv("C:\\Python main\\practise\\csv files\\3.Rev from countries.csv")
countries_df.head(10)
```

```python
fig = plt.figure()

plt.bar(countries_df['country'],countries_df['total_payment'])
spacing = 0.500
fig.subplots_adjust(bottom=spacing)
plt.xticks(rotation=45)

plt.xlabel('Countries')
plt.ylabel('Total revenue')
plt.title(' Total revenue from different countries')

plt.show()
```
